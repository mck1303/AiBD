```python
#Zadanie 3

import numpy as np
from matplotlib import pyplot as plt

def fun(x):
    return x**2+5

w1=np.linspace(-1,1,200)
w2=np.linspace(-6,6,1200)
w3=np.linspace(0,5,500)

f1=np.zeros(len(w1))
f2=np.zeros(len(w2))
f3=np.zeros(len(w3))
for i in range (len(w1)):
    f1[i]=fun(w1[i])
for i in range (len(w2)):
    f2[i]=fun(w2[i])
for i in range (len(w3)):
    f3[i]=fun(w3[i])
    
plt.figure(1)
z1, =plt.plot(w1,f1,label="x^2 +5")
plt.legend(handles=[z1], loc='lower right')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Wykres funkcji f(x)=x^2 +5 w przedziale (-1,1)")
plt.show()

plt.figure(2)
z2, =plt.plot(w2,f2,label="x^2 +5")
plt.legend(handles=[z2], loc='lower right')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Wykres funkcji f(x)=x^2 +5 w przedziale (-6,6)")
plt.show()

plt.figure(3)
z3, =plt.plot(w3,f3,label="x^2 +5")
plt.legend(handles=[z3], loc='lower right')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Wykres funkcji f(x)=x^2 +5 w przedziale (0,5)")
plt.show()
```
